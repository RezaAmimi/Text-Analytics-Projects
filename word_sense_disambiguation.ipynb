```python
from lxml import etree, objectify


# Read the dictionary file.
Parser = objectify.makeparser(recover=True)
Tree = objectify.fromstring(''.join(open('dictionary.xml').readlines()), Parser)
```

```python
# function: 
def getSenses(word, pos):
    global Tree
    item = Tree.xpath("//lexelt[@item='%s.%s']" % (word, pos))    
    senses = []
    for sense in item[0].getchildren():
        senses.append(dict(zip(sense.keys(), sense.values())))
    return senses
```

```python
# function:
def getSense(word, pos, sense_id):
    global Tree
    sense = Tree.xpath("//lexelt[@item='%s.%s']/sense[@id='%d']" % (word, pos, sense_id))
    return dict(zip(sense[0].keys(), sense[0].values()))
```

```python
def dataframe_laundrmat(dirty_df):
  column_names = ["target_word", "position", "sense", "previous_context", "target_word2", "post_context"]
  temp_df = pd.DataFrame(columns = column_names)

  for index, row in tqdm( dirty_df.iterrows()):
      string0 = row[0]
      string1 = row[1]
      string2 = row[2]

      string0_split = string0.split(".")
      string2_split = string2.split("%%")

      diict_2b_df = {'target_word':string0_split[0], "position":string0_split[1], "sense": row[1], "previous_context":string2_split[0].strip(), "target_word2":string2_split[1], "post_context":string2_split[2].strip()}
      temp_df = temp_df.append(diict_2b_df, ignore_index = True)

      # print(row[2])
      # break

  return temp_df
```

```python
#remmoving punctuation
import string
def remove_punctuation(context_data):
  y = context_data.translate(str.maketrans('', '', string.punctuation))
  return y
```

```python
#formating the data in the desiered format

def dataframe_laundrmat_2 (dirty_df, n_gram):
  column_names = ["target_word", "sense", 'context_vector']
  temp_df = pd.DataFrame(columns = column_names)

  for index, row in tqdm(dirty_df.iterrows()):
      string0 = remove_punctuation(expandContractions(row[3]))
      string1 = remove_punctuation(expandContractions(row[5]))

      string0_list=[]
      for i in string0.split(" "):
        if i != '':
          string0_list.append(i)

      string1_list=[]
      for i in string1.split(" "):
        if i != '':
          string1_list.append(i)


      temp_list=[]
      for l in range(-n_gram,0):
        if len(string0_list) >= n_gram:
          temp_list.append(string0_list[l])
      
      temp_list2=[]
      
      for m in range(0,n_gram):
        if len(string1_list) >= n_gram:
          temp_list2.append(string1_list[m])          

      mergedlist = []
      mergedlist.extend(temp_list)
      mergedlist.extend(temp_list2)

      diict_2b_df = {'target_word':row[0], "sense": row[2],"context_vector":mergedlist }
      temp_df = temp_df.append(diict_2b_df, ignore_index = True)

  return temp_df
```

```python
#getting the vocab and count from the dataset with all the colums 

def vocab (word_context_vectors):
  vocab = {}
  vocb_given_sense = {}
  sense_count = {}
  for i in range(len(word_context_vectors)):
    context_vector_list = word_context_vectors['context_vector'][i]
    for w in context_vector_list:
      if w not in vocab.keys():
        vocab[w] = 1
      else:
        vocab[w] += 1

  for i in vocab.keys():
    if i not in vocb_given_sense.keys():
      temp_dict = {}
      for j in unique_senses:
        temp_dict[j]=0 
      vocb_given_sense[i] = temp_dict
  for k in range(len(word_context_vectors)):
    for m in word_context_vectors['context_vector'][k]:
      vocb_given_sense[m][word_context_vectors['sense'][k]] += 1

  for i in unique_senses:
    sense_count[i]=0   

  for i in range(len(word_context_vectors)):
    context_vector_list_2 = word_context_vectors['context_vector'][i]
    # if word_context_vectors['sense'][i] not in sense_count.keys():
    #   sense_count[word_context_vectors['sense'][i]] = len(context_vector_list_2)
    # else: 
    sense_count[word_context_vectors['sense'][i]] += len(context_vector_list_2)

  return vocab ,vocb_given_sense,sense_count
```

```python
import numpy as np
def model_train_per_target_word (data_set, target_word,lam=0):
  column_names1 = ['target_word','vocab','sense','vocab_count','p(f/s)','p(s)']
  train_model = pd.DataFrame(columns= column_names1)
 
  #getting the required dataset for the word
  column_names2 = ["target_word", "sense", 'context_vector']
  data_set_target_word = pd.DataFrame(columns = column_names2)
  for i in range(len(data_set)):
    if data_set['target_word'][i] == target_word:
     x = pd.DataFrame(train_data_clean_2[i:i+1])
     data_set_target_word=data_set_target_word.append(x,ignore_index=True)
  vocab_dict, vocab_dict_given_sense, count_sense = vocab(data_set_target_word)

  for key, values in vocab_dict_given_sense.items():
    for key_val, val_val in values.items():
      target_w = target_word
      v = key
      s = key_val
      vc = val_val+lam
      if count_sense[key_val] == 0 and lam !=0:
         pf= lam/(lam*len(vocab_dict))
      elif count_sense[key_val] == 0 and lam ==0:
        pf= 0
      else:
        pf= (val_val+lam)/(count_sense[key_val]+lam*len(vocab_dict))
      ps = count_sense[key_val]/sum(count_sense.values())
      z= pd.DataFrame([[target_w,v,s,vc,pf,ps]], columns=column_names1)
      train_model = pd.concat([train_model,z],ignore_index=True)
  
  return train_model ,count_sense

      
```
